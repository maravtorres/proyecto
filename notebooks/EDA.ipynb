# Análisis de vehículos

Aplicación web interactiva creada con Streamlit para explorar datos de anuncios de coches en EE.UU.

## Funcionalidades
- Histograma del odómetro
- Gráfico de dispersión precio vs odómetro

## Tecnologías
- Python
- pandas
- plotly-express
- streamlit

## Ejecutar app
streamlit run app.py
